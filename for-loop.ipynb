```python
for i in range(1,6):
    for j in range(1,i+1):
        print(j, end=" ")
    print()
```

```python
for i in range(1,6):
    for j in range(1,i+1):
        print("*", end=" ")
    print()
```

```python
n = 5
for i in range(n):
    for j in range(n - i - 1):
        print(" ", end="")
    for k in range(2 * i + 1):
        print("*", end="")
    print()
```

```python
num = 1
for i in range(1, 6):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```
