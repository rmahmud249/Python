#String Method

```python
string="Welcome to my home town Barishal Barishal!!"

process_string=string.lower() #Coverting lowercase
process_string_1=string.upper()

print(len(string))
print("barishal" in process_string)
print("BARISHAL" in process_string_1)
print("Rasel" in string)


indx=process_string.find("barishal") # starting index 
print(indx)
indx_1=process_string.rfind("barishal")
print(indx_1)


count=process_string.count("barishal")#count containe
print(count)

#Replace a Substring
replace_string=string.replace("Barishal","Dhaka")
print(replace_string)
```
