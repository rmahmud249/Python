```python
def square_addition(a,b):
    a=a**2
    b=b**2
    return a+b

ans=square_addition(10,20)+100
print(ans)
```

```python
def s_addition(*args):
    summation = 0
    for i in args:
        i = i ** 2
        summation += i
    return summation


ans = s_addition(10, 20, 30, 40, 50)
print(ans)
```
